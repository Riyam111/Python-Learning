## Univariate Analysis---Single variable

```python
df['Genre']
```

```python
df['Genre'].unique()
```

```python
Genre_count=df['Genre'].value_counts()
```

```python
Genre_count
```

```python
#Bar plot for catgorical data (data which contain string or text)
x_bar=Genre_count.index
y_bar=Genre_count.values
```

```python
x_bar
```

```python
y_bar
```

```python
plt.bar(x_bar,y_bar)
```

### here you see bar chart but x acis is overlap with each other so it will correct by three method 1.rotate,2.change the fontsize decrese of fig,3.increse figure size

```python
plt.figure(figsize=(12,8))
plt.bar(x_bar,y_bar)
```

```python
plt.figure(figsize=(12,8))
plt.xticks(rotation=90,fontsize=10)
plt.bar(x_bar,y_bar,width=0.4)
```
